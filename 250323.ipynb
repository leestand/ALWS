```python
import pandas as pd
import requests
```

```python
name = '송화산시도삭면'  # 식당명
url = f'https://map.naver.com/p/api/search/allSearch?query={name}&type=all&searchCoord=127.09343900302565%3B37.548116747602194&boundary='
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
    "Referer": "https://map.naver.com"
}
response = requests.get(url, headers=headers)
```

```python
response.status_code
```

```python
response.json()['result']['place']['list']
```

```python
columns = [
    'name', 'category', 'categoryPath', 'rcode', 'address', 'roadAddress', 'reviewCount', 'placeReviewCount', 'x', 'y', 'menuInfo', 'microReview'
]

pd.DataFrame(response.json()['result']['place']['list'])[columns]
```

```python
pd.DataFrame(response.json()['result']['place']['list'])[columns][['name', 'menuInfo']]
```
